```python
##### Matplotlib을 활용한 시각화

# Matplotlib : 파이썬의 시각화 도구
# Numpy array를 기반으로 만들어진 다중 플랫폼 데이터 시각화 라이브러리로, 광범위한 SciPy 스택과 함께 작업하기 위해 설계됐다.
# Matplotlib은 다양한 운영체제와 그래픽 백엔드에서 잘 동작하는 특징으로 사용자층이 많아졌다.

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.__version__
```

```python
plt.style.use('classic')        # 여러 미적 스타일 중 가장 전형적인 classic 스타일을 설정함.
```

```python
### show()를 사용할 것인가, 말 것인가 - 플롯 표현 방법

# Matplotlib은 스크립트, IPython 터미널, IPython 노트북에서 사용할 수 있다. (여기선 노트북)
%matplotlib inline
# 이것을 통해 노트북에 플롯의 정적 이미지를 삽입할 수 있다.
# 이 명령어를 실행하면(커널/세션 당 1번만) 플롯을 생성하는 노트북 내의 셀이 결과 그래픽의 PNG  이미지를 삽입할 것이다.

x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
### 그림을 파일로 저장하기

# 위의 이미지를 savefig()를 통해 저장할 수 있다.
fig.savefig('my_figure.png')
# 이러면 현재 작업 디렉터리에 my_figure.png파일이 생성된다.
```

```python
##### 간단한 라인 플롯 (p261)

# 앞으로 나오는 예제에서 함수를 플로팅하고 임포팅하기 위한 노트북부터 설정
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# 모든 matplotlib 플롯은 그림(figure)과 축(axes)을 만드는 것으로 시작한다
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x));
```

```python
# pylab 인터페이스를 사용해 그림과 축이 백그라운드에 만들어지게 할 수 있다.
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
### 플롯 수정하기
# (1) 선 색상, (2) 선 스타일
plt.plot(x, np.sin(x), color='red', linestyle='--')
```

```python
# (3) 축 경계
plt.plot(x, np.sin(x))
plt.xlim(0,4)
plt.ylim(-1.5,1.5);
```

```python
plt.plot(x, np.sin(x))
plt.axis([0,4,-1.5,1.5]);     # plt.axis는 x축 최소최대값과 y축 최소 최대값을 한번에 지정해준다.
```

```python
# (4) 플롯에 레이블 붙이기
plt.plot(x, np.sin(x))
plt.title("plot main title")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
# (5) 플롯에 범례 삽입하기
plt.plot(x, np.sin(x), color='green', linestyle='-', label='sin(x)')
plt.plot(x, np.cos(x), color='red', linestyle='--', label='cos(x)')
plt.axis('equal')
plt.legend();
```

```python
##### 간단한 산점도 (p271)

x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'o',color='black');
```

```python
plt.plot(x,y,'-p',color='gray',markersize=15,linewidth=4);
```

```python
plt.scatter(x,y,marker='o');
```

```python
# plt.plot과의 주요 차이점으로 plt.scatter의 경우 각 점의 속성을 개별적으로 제어하는 산점도를 만들 수 있다.

from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2, s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
